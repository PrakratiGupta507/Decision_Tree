```python
import numpy as np
from scipy.io import loadmat
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset2 = loadmat('dataset_2.mat')
samples, label = dataset2['samples'], dataset2['labels']
df = pd.DataFrame(list(samples))
label = pd.DataFrame(label[0])
df['label'] = label
```

```python
x=df[0]
y=df[1]
z=df['label']
```

```python
data = pd.DataFrame({"X_Value": x, "Y_Value": y, "Category": z})
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(data=data, x="X_Value", y="Y_Value", hue="Category",palette="deep")
```
