```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
data = loadmat('dataset_1.mat')
```

```python
imgdata, target = data['samples'], data['labels'][0]
```

```python
space=[]
sudo=[]
length=target.shape[0]
for i in range(10):
    sudo = []
    count = 0
    for j in range(length):
        if(target[j] == i):
            sudo.append(imgdata[j])
            count+=1
        if count == 10:
            break
    space.append(sudo)
```

```python
space = np.array(space)
```

```python
w=10
h=10
fig=plt.figure(figsize=(10,10))
columns = 10
rows = 10
counter=1
for i in range(10):
    for j in range(10):
        fig.add_subplot(10,10,counter)
        counter+=1
        plt.imshow(space[i][j])
plt.show()
```
